## Map

```python
def mul(x):     return x*2
l = [1,2,3]
res = list(map(mul , l))
print(res)
```

```python
nums = [1,2,3,4]
sqnums = list(map(lambda x : x**2 , nums ))

print(sqnums)
```

```python
rate = [70 , 66, 50 , 40]
currency = [200 , 100, 50 , 20]

res = list(map(lambda x,y : x*y , rate , currency))
print(res)
```

```python
metal = ["gold" , "silver"]
res = list(map(str.capitalize, metal))
print(res)
```

```python
res = list(map(lambda x:x[0] , metal))
r = "".join(res)
print(r)
```

## Filter

```python
def even(x):
    if x%2 == 0 : return True
    else : return False

l = [1,2,3,4,5,6]
res = list(filter(even , l))
print(res)
```

```python
age = [25 , 65 , 89 , 45 , 23]

res = list(filter(lambda x: x>25 , age))
print(res)
```

```python
data = [0 , "hello" , 42 , None , "world" , "false" , False , 42]

falsywords = [0 , "" , False , None]

print(list(filter(lambda x: True if x not in falsywords else False , data )))
```

```python
original = ['ate', 'eat', 'abc', 'xyz', 'bca', 'rat', 'art', 'can', 'yzx', 'far']

anagram_dict = {}

for word in original:
    
    sorted_key = "".join(sorted(word))
    
    
    if sorted_key not in anagram_dict:
        anagram_dict[sorted_key] = []
    anagram_dict[sorted_key].append(word)


result = list(anagram_dict.values())
print(anagram_dict)
print(list(anagram_dict.items()))

print(result)
```
