```python
class Graph:
    def __init__(self, n):
        self.n = n
        self.g = [[] for _ in range(n)]

    def addEdge(self , u , v ):
        self.g[u].append(v)
        self.g[v].append(u)

    def dfs(self, visited, s):
        res = []
        
        self.df1(s, visited, res)
        return res

   
    def df1(self, s, visited, res):
       
        res.append(s)
        visited[s] = True
        
       
        for x in self.g[s]:
            if not visited[x]: 
                
                self.df1(x, visited, res)


obj = Graph(5)
print(obj.g)

obj.addEdge(0, 1)
obj.addEdge(1, 2)
obj.addEdge(2, 3)
obj.addEdge(3, 4)

print(obj.g)

v = [False]*5
obj.dfs(v , 2)
```
