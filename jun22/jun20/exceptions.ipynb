```python
def operate(a , b):

    print("-------------------------------------------------------------------------------------------------------------------")
    try:
        print("Addition",a+b)
        print("Subtraction",a-b)
        print("Multiplication",a*b)
        print("DivisioN",a/b)
    except ZeroDivisionError :
        b=2
        print("Line 500 : Error Zero Division Not possible Going with Default Value which is divisible by 2 " , a/b)
    else:
        print("Line 200 : No Exceptions Are Genarated ")
    
    finally:
        print("Line 200 : Program  Completed")

        print("-------------------------------------------------------------------------------------------------------------------")

        
operate(2,0)
operate(20 , 5)
```

```python
def powers(a):
    try:
        print("-------------------------------------------------------------------------------------------------------------------")
        a = int(a)
        print("Square Of the Number" , a**2)
        print("Cube of the NUmber" , a**3)
    except ValueError:  
        print("Line 500 : Converting letters or words To Integer Is not Possible ")
    else:
        print("Line 200 : No Exceptions Are Genarated ")
    finally:
        print("Line 200 : Program  Completed")
        print("-------------------------------------------------------------------------------------------------------------------")

powers(2)
powers("red")
```

```python
def handle(a , b):
    try:
        print("-------------------------------------------------------------------------------------------------------------------")

        res = a/b

        print(f"--------------------------------------------- \n Inputs : {a} , {b} \n----------------------------------------------")
        print("Line 200: Division Result" , res)
    except TypeError:

        if type(a) != int or type(a) !=  float:
            print("first Input Is Invalid")
            print("Line 500: Both Inputs Should Be numbers Only ")

        if type(b) != int or type(b) !=  float:
            print("Second Input Is Invalid")
            print("Line 500: Both Inputs Should Be numbers Only ")


    except ZeroDivisionError:
        print("Line 500: Division BY Zero Is Not Possible")


    else:
        print("Line 200 : No Exceptions Are Genarated ")
    finally:
         print("Line 200 : Program  Completed")
         print("-------------------------------------------------------------------------------------------------------------------")


handle("200" , 8)
handle(8 , "20")
handle(20 , 2)
handle(20.9 , 2.6)
handle(20 , 0)
```

```python
def operations(l , idx):
    try:
        print("-------------------------------------------------------------------------------------------------------------------")

        print(f"Sum: {sum(l)}\tAverage: {sum(l)/len(l)}\tMinimum : {min(l)}\t Maximum :{max(l)}")
        print("Element at Given Indes" , l[idx])
        

    except IndexError:
        print("Line 500 : Index Is Out OF Range ; Original List ", l ,"Valid Length:",len(l))
        print("-------------------------------------------------------------------------------------------------------------------")

    else :
        print("Line 200 : Program Completed. Without Exceptions")
        print("-------------------------------------------------------------------------------------------------------------------")


l = [0 , 1, 2, 3 , 4, 5, 6]
operations(l , 2)
operations(l , 9)
```

```python
def file_handle(file_path):
    try:
        with open(file_path, 'r') as f:
            l = f.readlines() 

        # print(l)
        print(len(l))
        
    except FileNotFoundError:
        print(f"Error: The file at '{file_path}' does not exist.")
        print("Creating the file, writing default content, and recalling function...")
        
       
        with open(file_path, 'w') as f:
            f.write("Line 1.\nLine 2\n")
            
        file_handle(file_path)  

file_handle('./file.txt')
file_handle('./a.txt')
```

```python
class InvalidAmountException(Exception):
    pass

class Bank:
    id=0
    def __init__(self,name1):
        Bank.id+=1
        self.name=name1
        self.acc_no=Bank.id
        self.balance=0
    def deposit(self,amount):
        self.balance=amount+self.balance
        print("Sucess")
    def withdrawl(self,a):
        try:
            if self.balance<a:
                raise InvalidAmountException
            else:
                self.balance-=a
        except InvalidAmountException:
            print("Invalid Amount")
    def checkBalance(self):
        print("Balance Amount : ",self.balance)
    def display(self):
        print("Name",self.name)
        print("Acc No",self.acc_no)
name=input("Enter Name : ")
b1=Bank(name)
while True:
    print("1.Deposit")
    print("2.WithDrawl")
    print("3.ChecKBalance")
    print("4.Display")
    print("5.Exit")
    ch=int(input())
    match ch:
        case 1:
            n=int(input("Enter Amount to Deposit : "))
            b1.deposit(n)
        case 2:
            n=int(input("Enter Amount to WithDrawl"))
            b1.withdrawl(n)
        case 3:
            b1.checkBalance()
        case 4:
            b1.display()
        case 5:
            exit()
        case _:
            print("Invalid Input...")
```
