```python
class DisConnectedGraphDictionary:
    def __init__(self, n):
        self.n = n
        self.g = {i: [] for i in range(n)}

    def addEdge(self, u, v):
        self.g[u].append(v)
        self.g[v].append(u)

    def display(self):
        print(self.g)
 
    def dfs1(self, s, visited, li):
        li.append(s)
        visited[s] = True
        
        # Iterates through the list of neighbors in the dictionary
        for i in self.g[s]:
            if not visited[i]:
                self.dfs1(i, visited, li)

    def dfs(self,s=4):
        la = []
        # Create a dictionary to track visited status for every node key in self.g
        visited = {node: False for node in self.g}
        
        # Iterate over all node keys to handle disconnected components
        for node in self.g:
            if not visited[node]:
                self.dfs1(node, visited, la)
                
        return la

obj = DisConnectedGraphDictionary(5)
print(obj.g)
```

```python
class Graph:
    def __init__(self, g):
        self.n = len(g)
        self.g = g
    
    def addEdge(self, u, v):
        self.g[u].append(v)
        self.g[v].append(u)

    def display(self):
        
        return self.g

    def getComponents(self):
        visited = [False] * self.n
        res = []
        for i in range(self.n):
            if not visited[i]:
                dummy = []
                self.dfs(i, visited, dummy)  
                res.append(dummy)
        return res

   
    def dfs(self, s, visited, dummy):
        visited[s] = True
        dummy.append(s)
        
        for neighbor in self.g[s]:
            if not visited[neighbor]:
                self.dfs(neighbor, visited, dummy)
   

g1 = Graph([[1, 2], [0, 3, 2], [0, 1], [1]])
g2 = Graph([[1, 2], [0, 3, 2], [0, 1], [1], []])


print(g1.display(),"Components" ,g1.getComponents() , "Length of Components" , len(g1.getComponents()) )

print(g2.display(),"Components" ,g2.getComponents() , "Length of Components" , len(g2.getComponents()) )
```
