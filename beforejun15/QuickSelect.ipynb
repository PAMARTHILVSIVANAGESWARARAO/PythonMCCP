```python
def quick_select(arr, left, right):
    pivot = arr[right]
    ploc = left  
    for j in range(left, right):
        if arr[j] < pivot:
            arr[ploc], arr[j] = arr[j], arr[ploc]
            ploc += 1

    arr[ploc], arr[right] = arr[right], arr[ploc]
    return ploc


def kth_smallest(arr, left, right, k): 
    while left <= right:
        p = quick_select(arr, left, right)

        if p == k:
            return arr[p]
        elif p < k:
            left = p + 1
        else:
            right = p - 1

    return None

arr = [10 , 20  , 30 , 40 , 50 , 60 , 70]
k = 3

ans = kth_smallest(arr, 0, len(arr) - 1, k - 1)
print(ans)
```
