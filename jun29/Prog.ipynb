```python
class DFU:
    def __init__(self, n):
        self.n = n 
        self.l = [i for i in range(self.n)]

    def find(self, i):
       
        if self.l[i] == i:
            return i
        
        return self.find(self.l[i])

    def union(self, u, v):
        i = self.find(u)
        j = self.find(v)

        if i != j:
            
            self.l[i] = j
```

```python
class Graph:
    def __init__(self, n):
        self.n = n 
        self.components = n  
        self.l = [x for x in range(self.n)]

        self.graph = {i: [] for i in range(self.n)}

    def find(self, i):
        if self.l[i] == i:  
            return i
        
        self.l[i] = self.find(self.l[i])
        return self.l[i]

    def union(self, u, v):
        self.graph[u].append(v)
        self.graph[v].append(u)

        i = self.find(u)
        j = self.find(v)

        if i != j:
            self.l[i] = j 
            self.components -= 1  

    def getComponents(self):
        groups = {}
        
        for i in range(self.n):
            root = self.find(i)
            
            
            if root not in groups:
                groups[root] = []
                
            groups[root].append(i)
            
        return groups

    def getGraph(self):
        return self.graph
```

```python
def process_grid(grid):
    # 1. Convert string elements to integers in-place
    for r in range(len(grid)):
        for c in range(len(grid[0])):
            grid[r][c] = int(grid[r][c])

    # 2. Get 1-based column indices for every '1' in each row
    # Filter out empty rows (rows with no 1s) so they don't cause IndexError later
    converted = [[c + 1 for c in range(len(row)) if row[c] == 1] for row in grid]
    converted = [r for r in converted if r] 

    # 3. Sort each individual row's indices (just in case they aren't sorted)
    for row in converted:
        row.sort()

    # 4. Remove duplicates based on the FIRST index value
    seen_first_indices = set()
    unique_rows = []
    
    # Sort the entire list of rows by their first element before filtering
    converted.sort(key=lambda x: x[0])

    for row in converted:
        first_idx = row[0]
        if first_idx not in seen_first_indices:
            seen_first_indices.add(first_idx)
            unique_rows.append(row)
            
    print(unique_rows)

# Test Cases
grid1 = [
  ["1","1","0","0","0"], # Indices: [1, 2] -> First idx: 1
  ["1","1","0","0","0"], # Indices: [1, 2] -> Duplicate first idx (1), will be removed
  ["0","0","1","0","0"], # Indices: [3]    -> First idx: 3
  ["0","0","0","1","1"]  # Indices: [4, 5] -> First idx: 4
]

grid2 = [
  ["1","1","1","1","0"], # Indices: [1, 2, 3, 4] -> First idx: 1
  ["1","1","0","1","0"], # Indices: [1, 2, 4]    -> Duplicate first idx (1), will be removed
  ["1","1","0","0","0"], # Indices: [1, 2]       -> Duplicate first idx (1), will be removed
  ["0","0","0","0","0"]  # Empty row -> Filtered out
]

print("Grid 1 Output:")
process_grid(grid1) 

print("\nGrid 2 Output:")
process_grid(grid2)
```
