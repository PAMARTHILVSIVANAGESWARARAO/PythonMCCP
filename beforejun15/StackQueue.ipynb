```python

def sort_tuples_by_digits(tuples_list):
    def digits_in_tuple(t):
        return sum(len(str(x)) for x in t)

    return sorted(tuples_list, key=digits_in_tuple)


data = [(3, 4, 6, 723), (1, 2), (124, 324, 72)]
print(sort_tuples_by_digits(data))  



```

```python
def tuples_all_divisible_by_k(tuples_list, k):
    return [t for t in tuples_list if all(x % k == 0 for x in t)]

data = [(3, 4, 6, 723), (1, 2), (124, 324, 72)]
print(tuples_all_divisible_by_k(data, 2))

```

```python
def remove_tuples_with_length_k(tuples_list, k):
    return [t for t in tuples_list if len(t) != k]


data = [(3, 4, 6, 723), (1, 2), (124, 324, 72)]
print(remove_tuples_with_length_k(data, 2))

```

```python
def sort_tuples_by_second(tuples_list):
    return sorted(tuples_list, key=lambda t: t[1])

data = [(3, 10, 6), (1, 2), (124, 324, 72)]
print(sort_tuples_by_second(data)) 
```
