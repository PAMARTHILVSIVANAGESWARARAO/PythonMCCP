```python
obj = LinkedList.LinkedList()
```

```python
menu_functions = obj.__all__
```

```python
print(menu_functions)
```

```python
while(True):
    for i in range(len(menu_functions)-2):
        print(f"{i+1}.{menu_functions[i]}")

        
    choice = int(input("Enter Your Choice"))
    

    match choice:
        case 1:
            x = int(input("Enter Element to Insert"))
            obj.add(x)
            obj.traverse()

        case 2:
            x = int(input("Enter Element to Insert at First"))
            obj.add_first(x)
            obj.traverse()

        case 3:
            x = int(input("Enter Element to Insert at Last"))
            obj.add_last(x)
            obj.traverse()

        case 4:
            data , pos = map(int , input("Enter (data , position )").split())
            obj.add_at_position(data , pos)
            obj.traverse()

        case 5:
            obj.traverse()

        case 6:
            obj.delete_head()
            obj.traverse()

        case 7:
            obj.delete_end()
            obj.traverse()

        case 8:
            key = int(input("Enter Element to Search:"))
            print("Exist" if obj.search(key) == True else "Not exist")

        case 9:
            key = int(input("Enter Element to Recursive Search:"))
            print("Exist" if obj.search(key) == True else  "Not exist")

        case _:
            break
        

```

```python
from threading import Thread
from time import sleep

def f1():   print("From Function-1");sleep(1)
def f2():   print("From Function-2");sleep(2)

t1,t2 = Thread(f1()),Thread(f2())
t1.start()
t1.join()
t2.start()
t2.join()
```

```python
def func(name): 
    if name=='A':   
        for i in range(0 , 10 , 2):    print(i);sleep(1)

    else:   
        for i in range(1 , 10 , 2):    print(i);sleep(1)

t1,t2 = Thread(func('A')),Thread(func('B'))
t1.start()
t2.start()
t1.join()
t2.join()
```

```python
from concurrent.futures import ThreadPoolExecutor
def mul(x): return 2*x
l=[10,20,30,40]
e = ThreadPoolExecutor(max_workers=  2)


res = e.map(mul , l)
print(list(res))
```
