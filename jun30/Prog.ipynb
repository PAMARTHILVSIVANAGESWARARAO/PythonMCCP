```python
from collections import deque
from typing import List

class Graph:
    def __init__(self, n, g):
        self.n = n
        self.g = g

    def bfs(self, start_node, colors) -> bool:
        
        queue = deque([start_node])
        colors[start_node] = 0  

        while queue:
            node = queue.popleft()

            for neighbor in self.g[node]:
                
                if colors[neighbor] == -1:
                    colors[neighbor] = 1 - colors[node]
                    queue.append(neighbor)
                
                elif colors[neighbor] == colors[node]:
                    return False
        
        return True

    def greedy_coloring(self) -> bool:
        
        colors = [-1] * self.n  

        for i in range(self.n):
            if colors[i] == -1:  
                if not self.bfs(i, colors):
                    return False
        
        return True

class Solution:
    def isBipartite(self, graph: List[List[int]]) -> bool:
        obj = Graph(len(graph), graph)
        return obj.greedy_coloring()
```

```python
class Graph:
    def __init__(self, g):
       
        self.g = g
        
        self.n = len(self.g)

    def detect_cycle(self):
        state = {node: 0 for node in self.g}
        
        def dfs(node):
            state[node] = 1
            
            for neighbor in self.g.get(node, []):
                if state[neighbor] == 1:
                    return True
                elif state[neighbor] == 0:
                    if dfs(neighbor):
                        return True
            
            state[node] = 2
            return False

        for node in self.g:
            if state[node] == 0:
                if dfs(node):
                    return True
                    
        return False
```

```python
def dfs(img, x, y, oldColor, newColor):
    if (x < 0 or x >= len(img) or
        y < 0 or y >= len(img[0]) or
        img[x][y] != oldColor):
        return

   
    img[x][y] = newColor

    
    dfs(img, x + 1, y, oldColor, newColor)
    dfs(img, x - 1, y, oldColor, newColor)
    dfs(img, x, y + 1, oldColor, newColor)
    dfs(img, x, y - 1, oldColor, newColor)


def floodFill(img, sr, sc, newColor):
    
   
    if img[sr][sc] == newColor:
        return img

    
    oldColor = img[sr][sc]
    dfs(img, sr, sc, oldColor, newColor)

    return img
  
    
if __name__ == "__main__":
    img = [
        [1, 1, 1, 0],
        [0, 1, 1, 1],
        [1, 0, 1, 1]
    ]
    
    sr, sc = 1, 2
    newColor = 2
    
    result = floodFill(img, sr, sc, newColor)
    
    for row in result:
        print(*row)
```

```python
class DFU:
    def __init__(self, n):
        self.n = n 
        # Create elements from 0 to n. 
        self.l = [i for i in range(self.n + 1)]
        self.s = [1] * (self.n + 1)

    def find(self, i):
        if self.l[i] == i:
            return i
        # Path compression: update self.l[i] directly
        self.l[i] = self.find(self.l[i])
        return self.l[i]

    def union(self, u, v):
        i = self.find(u)
        j = self.find(v)

        if i != j:
            # Attach root i to root j
            self.l[i] = j
            # The new root j absorbs the size of group i
            self.s[j] += self.s[i]


n = int(input("Enter No of Companies: "))
obj = DFU(n)

while True:
    print("""
            1. C x Y (Merge Companies)
            2. Q x (Show Number of Merged Groups)
            3. Show Assignments & Sizes
            4. Exit
        """)
    try:
        choice = int(input("Enter Choice: "))
    except ValueError:
        break
        
    match choice:
        case 1:
            print("Merge Companies (Starting from 1)")
            x = int(input("Enter First Company Number: "))
            y = int(input("Enter Second Company Number: "))
            
            # Boundary check to prevent IndexError
            if 1 <= x <= n and 1 <= y <= n:
                obj.union(x, y)
                print(f"Companies {x} and {y} merged successfully.")
            else:
                print("Invalid company numbers!")

        case 2:
            # Check roots from 1 to n (ignoring index 0)
            group_count = sum(1 for i in range(1, obj.n + 1) if obj.l[i] == i)
            print(f"Number of Independent Company Groups: {group_count}")
            
        case 3:
            # Flatten the tree structures before printing
            for i in range(1, obj.n + 1):
                obj.find(i)
            print(f"Group assignments (1 to {n}): {obj.l[1:]}")
            
            x = int(input("Enter Company Number to check total group size: "))
            if 1 <= x <= n:
                root = obj.find(x)
                print(f"Total companies merged with {x}: {obj.s[root]}")
            
        case _:
            print("Exiting...")
            break
```

```python
def cal(nums, target):
    nums = sorted(nums)
    res = []
    
    for i in range(len(nums) - 2):
        # Skip duplicate values for the first element to avoid identical triplets
        if i > 0 and nums[i] == nums[i - 1]:
            continue
            
        j = i + 1
        k = len(nums) - 1
        
        while j < k:
            s = nums[i] + nums[j] + nums[k]
            
            if s == target:
                res.append([nums[i], nums[j], nums[k]]) # Append values, not indices
                j += 1
                k -= 1
                
                # Skip duplicate values for the left pointer
                while j < k and nums[j] == nums[j - 1]:
                    j += 1
                # Skip duplicate values for the right pointer
                while j < k and nums[k] == nums[k + 1]:
                    k -= 1
                    
            elif s > target:
                k -= 1
            else:
                j += 1

    return res

# Testing your exact example:
print(cal([-1, 0, 1, 2, -1, -4], 0))
# Output: [[-1, -1, 2], [-1, 0, 1]]
```
