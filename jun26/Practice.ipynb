```python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None


class LinkedList:

    def __init__(self):
        self.head = None


    def getList(self):
        curr = self.head
        res = []
        while curr:
            res.append(curr.data)
            curr= curr.next
        return res

    def deleteLast(self , position=None):
        if position is None:
            position = self.size() - 1
        self.deleteAtPosition(position)

    def deleteFirst(self , position=None):
        if position is None:
            position = 0
        self.deleteAtPosition(position)

    def addFirst(self , data , position=0 ):
        self.addAtPosition(position , data)


    def addLast(self , data , position=None):
        if position is None:
            position = self.size()
        self.addAtPosition(position , data)

    def addAtPosition(self, position, new_data):

        x = Node(new_data)

        if position == 0:
            x.next = self.head
            self.head = x
            return

        curr = self.head

        for i in range(position - 1):
            if curr is None:
                raise IndexError("Position out of bounds")
            curr = curr.next

        if curr is None:
            raise IndexError("Position out of bounds")

        x.next = curr.next
        curr.next = x
        

    def deleteAtPosition(self, position):

        if self.head is None:
            raise IndexError("Delete from empty list")

        if position == 0:
            self.head = self.head.next
            return

        curr = self.head

        for i in range(position - 1):
            if curr.next is None:
                raise IndexError("Position out of bounds")
            curr = curr.next

        if curr.next is None:
            raise IndexError("Position out of bounds")

        curr.next = curr.next.next

    def traverse(self):
        curr = self.head
        while curr:
            print(curr.data, end="->")
            curr = curr.next
        print("None")

    def size(self):
        curr= self.head 
        count = 0
        while curr:
            count+=1
            curr = curr.next
        return count
```
