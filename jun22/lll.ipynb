```python
class Node:
    def __init__(self , data ):
        self.data = data 
        self.next = None 


# a = Node(10)   |10|
# b = Node(20)   |20|   
# c = Node(30)   |30|

# a.next = b    |10| | -> |20| |
# b.next = c    |20| | -> |30| |
# head = a      a is the begginig 
# print(head is a) #True
```

```python

a = Node(10)   
b = Node(20)      
c = Node(30)   

a.next = b        
b.next = c        
head = a      
print(head is a) #True
print(a.data)
print(head.data)
print(head.next.data)
print(head.next.next.data)
```

```python
def traverse(head):
    curr = head 

    

    while curr is not None:
        print(curr.data , " -> " , end="" )
        curr = curr.next
        

    print("None")

traverse(a)
traverse(b)
traverse(c)
```

```python
def asList(head):
    curr = head 
    res = []
    while curr is not None:
        res.append(curr.data)
        curr = curr.next
    return res

print(asList(a))
print(asList(b))
print(asList(c))
```

```python
class Node:
    def __init__(self , data):
        self.data = data 
        self.next = None 

```

```python
class LinkedList:
    

    def __init__(self):
        self.head = None


    def insert_at_begin(self , x):
        t = Node(x)
        t.next = self.head 
        self.head = t
        return self.head

    def insert_at_last(self , x):
        t = Node(x)

        curr = self.head 
        while curr.next !=None:
            curr = curr.next
        curr.next = t

        return self.head

    def insert_at_position(self , x , pos):
        t = Node(x)
        if pos == 0:
            return self.insert_at_begin(x)
        curr = self.head 
        for i in range(pos - 1):
            curr = curr.next
        t.next = curr.next
        curr.next = t

        return self.head

    def del_begin(self):
        self.head = self.head.next 
        return self.head

    def del_at_last(self):
        curr = self.head 

        while curr.next.next != None:
            curr = curr.next
        curr.next = None

        return self.head 
    def del_at_idx(self , pos):
        curr = self.head 

        for i in range(pos-1):
            curr = curr.next 
        curr.next = curr.next.next

        return self.head

    def traverse(self):
        curr = self.head

        while curr!=None:
            print(curr.data, " -> " , end=" ")
            curr = curr.next
```

```python
ll = LinkedList()

print("Insert at beginning:")
ll.insert_at_begin(30)
ll.insert_at_begin(20)
ll.insert_at_begin(10)
ll.traverse()
print("\n")

print("Insert at last:")
ll.insert_at_last(40)
ll.insert_at_last(50)
ll.traverse()
print("\n")

print("Insert 25 at position 2:")
ll.insert_at_position(25, 2)
ll.traverse()
print("\n")

print("Delete beginning:")
ll.del_begin()
ll.traverse()
print("\n")

print("Delete last:")
ll.del_at_last()
ll.traverse()
print("\n")

print("Delete node at index 2:")
ll.del_at_idx(2)
ll.traverse()
print()
```
