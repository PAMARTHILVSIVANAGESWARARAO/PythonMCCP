```python
class ArthemeticOperations:
    def __init__(self, x, y):
        self.x = x 
        self.y = y

        # print("line 200",self.x , self.y)

   
    def __add__(self, dobj):
 
        new_x = self.x + dobj.x
        new_y = self.y + dobj.y
        
      
        return ArthemeticOperations(new_x, new_y)


    def __sub__(self, dobj):
 
        new_x = self.x - dobj.x
        new_y = self.y - dobj.y
        
      
        return ArthemeticOperations(new_x, new_y)

    def __mul__(self, dobj):
 
        new_x = self.x * dobj.x
        new_y = self.y * dobj.y
        
      
        return ArthemeticOperations(new_x, new_y)


    def __str__(self):
        return f"({self.x}, {self.y})"


p1 = ArthemeticOperations(2, 4)
p2 = ArthemeticOperations(8, 6)


p3 = p1 + p2

print(f"p1: {p1}")
print(f"p2: {p2}")
print(f"p3 (p1 + p2): {p3}")
```
