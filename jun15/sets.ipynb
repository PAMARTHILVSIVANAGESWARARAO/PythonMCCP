```python
s = set()

while True:
    print("---------------------------------------")
    print("1. Add")
    print("2. Remove")
    print("3. Display")
    print("4. Exit")
    print("---------------------------------------")

    case = int(input("Enter your choice: "))

    match case:
        case 1:
            x = int(input("Enter Value to add: "))
            s.add(x)
            print("Set:", s)

        case 2:
            x = int(input("Enter Value to remove: "))
            s.discard(x)
            print("Set:", s)

        case 3:
            print(s if len(s) != 0 else "Set is Empty")

        case 4:
            break

        case _:
            print("Invalid choice. Try again.")
```

```python
# Static set
numbers = {10, 20, 30, 40, 50}

# Sum of elements
total = sum(numbers)

# Average of elements
average = total / len(numbers)

print("Set:", numbers)
print("Sum:", total)
print("Average:", average)
```

```python
# Static sets
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

# Union
union_set = set1.union(set2)

# Intersection
intersection_set = set1.intersection(set2)

print("Set 1:", set1)
print("Set 2:", set2)
print("Union:", union_set)
print("Intersection:", intersection_set)
```

```python
# Static sets
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}

# A - B
a_minus_b = A - B

# B - A
b_minus_a = B - A

# Symmetric Difference
sym_diff = A.symmetric_difference(B)

print("Set A:", A)
print("Set B:", B)
print("A - B:", a_minus_b)
print("B - A:", b_minus_a)
print("Symmetric Difference:", sym_diff)
```

```python
# Static set
s = {12, 45, 7, 23, 89, 34}

# Using built-in functions
max_builtin = max(s)
min_builtin = min(s)
range_builtin = max_builtin - min_builtin

print("Using Built-in Functions:")
print("Maximum =", max_builtin)
print("Minimum =", min_builtin)
print("Range =", range_builtin)

# Without using built-in max() and min()
s_list = list(s)
max_val = s_list[0]
min_val = s_list[0]

for num in s_list:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num

range_val = max_val - min_val

print("\nWithout Built-in Functions:")
print("Maximum =", max_val)
print("Minimum =", min_val)
print("Range =", range_val)
```

```python
# Static sets
A = {1, 2, 3}
B = {1, 2, 3, 4, 5}

# Subset
print("A is subset of B:", A.issubset(B))

# Superset
print("B is superset of A:", B.issuperset(A))

# Proper Subset
print("A is proper subset of B:", A < B)

# Proper Superset
print("B is proper superset of A:", B > A)
```

```python
# Static sets
A = {1, 2, 3}
B = {4, 5, 6}

# Check if sets are disjoint
print("A:", A)
print("B:", B)
print("Are A and B disjoint?", A.isdisjoint(B))
```

```python
# List with duplicates
lst = [1, 2, 2, 3, 4, 4, 5, 1]

# Using set
unique_list = list(set(lst))

print("Original List:", lst)
print("After Removing Duplicates (Using Set):", unique_list)


# List with duplicates
lst = [1, 2, 2, 3, 4, 4, 5, 1]

unique_list = []

for item in lst:
    if item not in unique_list:
        unique_list.append(item)

print("Original List:", lst)
print("After Removing Duplicates (Without Set):", unique_list)
```

```python
# List
lst = [1, 2, 3, 4, 5]

# Check uniqueness by comparing lengths
if len(lst) == len(set(lst)):
    print("List contains unique elements")
else:
    print("List contains duplicate elements")
```

```python
# Array of numbers
arr = [12, 18, 25, 35]

unique_prime_factors = set()

for num in arr:
    d = 2
    while num > 1:
        while num % d == 0:
            unique_prime_factors.add(d)
            num //= d
        d += 1

print("Unique Prime Factors in Array:", unique_prime_factors)
```

```python
import random

sample_names = ["Alice", "Bob", "Charlie", "David", "Emma", "Frank", "Grace", "Henry", "Ivy", "Jack"]
sample_jobs = ["Manager", "Developer", "Designer", "Analyst", "HR Specialist", "QA Engineer"]

n = int(input("Enter n (number of employees): "))

d = {}

for i in range(1, n + 1):
    
    name = random.choice(sample_names)
    job = random.choice(sample_jobs)
    
    salary = random.randint(40000, 120000)
    
    d[i] = {
        "name": name,
        "job": job,
        "salary": salary
    }


print("\nAll employee details (Automatically Generated):")
for emp_id, info in d.items():
    print(f"ID {emp_id}: {info}")


search_name = input("\nEnter employee name to display details: ")

found = False
for emp_id, info in d.items():
    if info["name"].lower() == search_name.lower():  
        print(f"\nEmployee details for '{search_name}':")
        print({emp_id: info})
        found = True
        break

if not found:
    print(f"No employee found with name '{search_name}'.")
```
