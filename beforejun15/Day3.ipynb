```python


def consecutive_sequence(l):
    sorted_l = sorted(l)
    max_count = 1
    current_count = 1
    for i in range(1, len(sorted_l)):
        if sorted_l[i] == sorted_l[i-1] + 1:
            current_count += 1
        else:
            max_count = max(max_count, current_count)
            current_count = 1
    max_count = max(max_count, current_count)
    print(l , "=>", max_count)
    
consecutive_sequence([100 , 4 , 900 , 1 , 3 , 2 , 9 , 11 , 12 , 10 , 15 , 13 , 16 , 14])
consecutive_sequence([42])
consecutive_sequence([100 , 4, 200 , 1 , 3 , 2])
consecutive_sequence([1 , 2 , 3, 4, 5])
consecutive_sequence([5 , 4, 3, 2, 1])
consecutive_sequence([7 , 7 , 7 , 7])
consecutive_sequence([-3 ,-1 , 0 , -2 , 3 ,1 , 20])
```

```python
import random as rand 

names = ['Rocky' , 'Bahubali' , "OG" , "Sahoo" ]
jobs = ["frontend dev" , "backend dev" , "full stack dev" , "ai dev "]
location = ["mumbai" , "pune" , "Patna"]


employee = []
for i in range( 1 , 10):
    l = []
    l.append(i)
    l.append(names[rand.randint(0,len(names)-1)])
    l.append(jobs[rand.randint( 0 , len(jobs)-1)])
    l.append(rand.randint(10000,90000))
    l.append(location[rand.randint(0 , len(location)-1)])
    l.append(rand.randint(1,5))

    employee.append(l)
print(employee)



```
